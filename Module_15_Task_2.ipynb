```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25)
]
```

```python
df = pd.DataFrame(prices, columns=["Month", "Price"])
df.set_index('Month')
```

```python
df["priceUSD"] = df["Price"]/4
df
```

```python
plt.plot(df["priceUSD"],linestyle="--", color="red")
plt.xlabel("month")
plt.ylabel("priceUSD")
plt.title("Price of goods (USD)")
plt.show()
```
